#### Regularization:
- Use the `diabetes` dataset from `sklearn.datasets`.
- Compare the performance (Mean Squared Error) of `LinearRegression`, `Ridge`, and `Lasso` models.
- Tune the `alpha` parameter for `Ridge` and `Lasso` using `GridSearchCV` with cross-validation to find the optimal regularization strength.

```python
import numpy as np
from sklearn.linear_model import LinearRegression, Ridge, Lasso
from sklearn.model_selection import GridSearchCV, train_test_split
from sklearn.metrics import mean_squared_error

from sklearn.datasets import load_diabetes
```

```python
# Load the diabetes dataset
diabetes = load_diabetes()

# Split data
X_train, X_test, y_train, y_test = train_test_split(diabetes.data, diabetes.target, test_size=0.2, random_state=0)
```

```python
# Linear regression
linear_reg = LinearRegression()
linear_reg.fit(X_train, y_train)
linear_pred = linear_reg.predict(X_test)
print(f'Linear Regression MSE: {mean_squared_error(y_test, linear_pred)}')

# Ridge regression + tune alpha param using GridSearchCV, 
ridge_reg = Ridge()
ridge_param_grid = {'alpha': [1e-10, 1e-9, 1e-8, 1e-7]}
ridge_gs = GridSearchCV(estimator=ridge_reg, param_grid=ridge_param_grid, cv=5) # default score R2
ridge_gs.fit(X_train, y_train)
ridge_pred = ridge_gs.predict(X_test)

print(f'Ridge MSE: {mean_squared_error(y_test, ridge_pred)}')
print(f'Ridge best alpha: {ridge_gs.best_params_}')
print(f'Ridge best score R2: {ridge_gs.best_score_}')

# Lasso regression + tune alpha param using GridSearchCV, 
lasso_reg = Lasso()
lasso_param_grid = {'alpha': [1e-7, 1e-5, 1e-3, 1]}
lasso_gs = GridSearchCV(estimator=lasso_reg, param_grid=lasso_param_grid, cv=5) # default score R2
lasso_gs.fit(X_train, y_train)
lasso_pred = lasso_gs.predict(X_test)

print(f'Lasso MSE: {mean_squared_error(y_test, lasso_pred)}')
print(f'Lasso best alpha: {lasso_gs.best_params_}')
print(f'Lasso best score R2: {lasso_gs.best_score_}')
```

#### Ensemble Methods:

- Use the `breast_cancer` dataset from `sklearn.datasets`.
- Compare the performance (F1 Score and AUC) of `DecisionTreeClassifier`, `RandomForestClassifier`, and `GradientBoostingClassifier`.
- Tune the hyperparameters of each classifier using `GridSearchCV` with cross-validation.

```python
import numpy as np
from sklearn.model_selection import GridSearchCV, train_test_split
from sklearn.tree import DecisionTreeClassifier
from sklearn.ensemble import RandomForestClassifier, GradientBoostingClassifier
from sklearn.metrics import f1_score, roc_auc_score

from sklearn.datasets import load_breast_cancer
```

```python
# Load the breast cancer dataset
breast_cancer = load_breast_cancer()

# Split data
X, y = breast_cancer.data, breast_cancer.target
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=0)
```

```python
# DecisionTreeClassifier
dt_clf = DecisionTreeClassifier() # default is 'gini'
dt_param_grid = {
    'max_depth': [2, 3, 5, 7],
    'min_samples_split': [4, 6, 8, 10],
    'min_samples_leaf': [2, 4, 6, 8],
}
dt_clf_gs = GridSearchCV(estimator=dt_clf, param_grid=dt_param_grid, cv=5, scoring='f1')
dt_clf_gs.fit(X_train, y_train)
dt_pred = dt_clf_gs.predict(X_test)
print('DT F1 score:', f1_score(y_test, dt_pred))
print('DT ROC AUC score:', roc_auc_score(y_test, dt_pred))
print('DT best params:', dt_clf_gs.best_params_)
```

```python
# RandomForestClassifier
rf_clf = RandomForestClassifier()
rf_param_grid = {
    'max_depth': [2, 3, 5, ],
    'min_samples_split': [8, 10, 20],
    'min_samples_leaf': [2, 4, 6],
}
rf_clf_gs = GridSearchCV(estimator=rf_clf, param_grid=rf_param_grid, cv=5, scoring='f1')
rf_clf_gs.fit(X_train, y_train)
rf_pred = rf_clf_gs.predict(X_test)
print('RF F1 score:', f1_score(y_test, rf_pred))
print('RF ROC AUC score:', roc_auc_score(y_test, rf_pred))
print('RF best params:', rf_clf_gs.best_params_)
```

```python
# GradientBoostingClassifier
gb_clf = GradientBoostingClassifier()
gb_param_grid = {
    'n_estimators': [50, 100, 200],       
    'learning_rate': [0.001, 0.01, 0.1]
}
gb_clf_gs = GridSearchCV(estimator=gb_clf, param_grid=gb_param_grid, cv=5, scoring='f1')
gb_clf_gs.fit(X_train, y_train)
gb_pred = gb_clf_gs.predict(X_test)
print('GB F1 score:', f1_score(y_test, gb_pred))
print('GB ROC AUC score:', roc_auc_score(y_test, gb_pred))
print('GB best params:', gb_clf_gs.best_params_)
```
